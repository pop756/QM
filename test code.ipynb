```python
class Drug:
    def __init__(self,data,target):
        Drug.x = data
        Drug.y = target
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem

def smile_to_RDkit(smiles_string,bits = 1024):
    # 예시 SMILES 문자열

    # SMILES 문자열을 RDKit 분자 객체로 변환
    molecule = Chem.MolFromSmiles(smiles_string)

    if molecule is not None:
        # 분자를 Morgan fingerprint로 변환 (다른 fingerprint 타입도 사용 가능)
        fingerprint = AllChem.GetMorganFingerprintAsBitVect(molecule, radius=2, nBits=bits)

        # fingerprint를 이진 벡터로 변환
        fingerprint_bits = list(fingerprint)

        # 결과 출력
        return fingerprint_bits

    else:
        print("error")
        return [1]*bits
```

```python
import csv
import numpy as np
f = open('names_smiles.csv', 'r', encoding='utf-8')
rdr = csv.reader(f)
test_samples = Drug([],[])
for line in rdr:
    test_samples.x.append(np.array(smile_to_RDkit(line[1])))
print(test_samples.x)
f = open('names_labels.csv', 'r', encoding='utf-8')
rda = csv.reader(f)
for line in rda:
    test_samples.y.append(int(line[1]))
print(test_samples.y)
```
